```python
import numpy as np
import pandas as pd
```

```python
BigFive = pd.read_csv("../results/BigFive.csv",index_col=['uid'])
flourishing = pd.read_csv("../results/flourishing.csv",index_col=['uid'])
loneliness = pd.read_csv("../results/loneliness.csv",index_col=['uid'])
panas = pd.read_csv("../results/panas.csv",index_col=['uid'])
stress  = pd.read_csv("../results/PerceivedStressScale.csv",index_col=['uid'])
phq9 = pd.read_csv("../results/PHQ-9.csv",index_col=['uid'])
vr12 = pd.read_csv("../results/vr_12.csv",index_col=['uid'])
```

```python
summary = pd.DataFrame({'uid': ['u{:02d}'.format(i) for i in range(60)]})
summary = summary.set_index('uid').join([BigFive, flourishing, loneliness, panas, stress, phq9, vr12])
summary.head()
```

```python
summary.to_csv("../results/survey_summary.csv")
```
