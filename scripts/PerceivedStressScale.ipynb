```python
data_path = '/Users/catherine/AI/UoA/T6/RA/code/StudentLife/data/dataset/survey/PerceivedStressScale.csv'
# define a mapping function
mapping_rule = [{
        'Never': int(0),
        'Almost never': int(1),
        'Sometime': int(2),
        'Fairly often': int(3),
        'Very often': int(4)
    }]
# Remove the row wtih NaN value
df, pre_df, post_df = pre_post_reader(data_path, mode='remove', mapping=mapping_rule,to_int=True)
```

```python
col_names = pre_df.columns[2:]
diff_df = get_pre_post_difference(pre_df, post_df, col_names,'PerceivedStressScale')
diff_df.head()
```

```python
# Mapping function
def mapping_status(score):
    if 0 <= score <= 13:
        return "Low stress"
    elif 14 <= score <= 26:
        return "Moderate stress"
    elif 27 <= score <= 40:
        return "High preceived stress"
    else:
        return "Unknown"

# Apply the mapping function to the scores
diff_df = get_status(diff_df, mapping_status, 'PerceivedStressScale')
diff_df.head()
```
